```python
# Setup device agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

### Get data

```python
import os
import zipfile
import requests

from pathlib import Path

# Setup path for dataset
data_path = Path("/data")
image_path = data_path / "pizza_steak_sushi"

# If the image folder doesn't exist
if image_path.is_dir():
  print(f"{image_path} exists.")
else:
  print(f"Did not find {image_path}, creating now...")
  image_path.mkdir(parents=True, exist_ok=True)

  # Download data
  with open(data_path / "pizza_steak_sushi.zip", "wb") as f:
    request = requests.get("https://github.com/mrdbourke/pytorch-deep-learning/raw/main/data/pizza_steak_sushi.zip")
    print("Downloading pizza, steak, sushi data...")
    f.write(request.content)

  with zipfile.ZipFile(data_path /"pizza_steak_sushi.zip", "r") as zip_ref:
    print("Unzipping dataset")
    zip_ref.extractall(image_path)

  os.remove(data_path /"pizza_steak_sushi.zip")

```

```python
# Setup directories for test and train sets
train_dir = image_path / "train"
test_dir = image_path / "test"
```

### Prepare data

```python
# Create a transforms pipeline
transform_manually = transforms.Compose([
                                        transforms.Resize(size=(224,224)),
                                        transforms.ToTensor(),
                                        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                             std=[0.229, 0.224, 0.225])
])
```

```python
# Create training and testing DataLoader's as well as get a list of class names
from going_modular.going_modular import data_setup


train_dataloader, test_dataloader, class_names = data_setup.create_dataloaders(train_dir=train_dir,
                                                                               test_dir=test_dir,
                                                                               transform=transform_manually,
                                                                               batch_size=32)
train_dataloader, test_dataloader, class_names
```

### Get and prepare a pretrained model

```python
# Setup the model with pretrained weights and send it to the target device
base_model = torchvision.models.efficientnet_b0(pretrained=True).to(device)
#model_0 # uncomment to output (it's very long)
```

```python
# Freeze all base layers in the "features" section of the model (the feature extractor) by setting requires_grad=False
for params in base_model.features.parameters():
  params.requires_grad= False
```

```python
# Set the manual seeds
torch.manual_seed(42)
torch.cuda.manual_seed(42)

# Get the length of class_names (one output unit for each class)
output_shape_for_feature = len(class_names)

# Recreate the classifier layer and seed it to the target device
base_model.classifier = nn.Sequential(
    torch.nn.Dropout(p=0.2, inplace=True),
    torch.nn.Linear(in_features=1280,
                    out_features=output_shape_for_feature,
                    bias=True)).to(device)
```

### Train model

```python
# Define loss and optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(base_model.parameters(), lr=0.001)
```

```python
from going_modular.going_modular import engine
# Set the random seeds
torch.manual_seed(42)
torch.cuda.manual_seed(42)

# Start the timer
from timeit import default_timer as timer
start_time = timer()

# Setup training and save the results
base_model_results = engine.train(model=base_model,
                                  train_dataloader=train_dataloader,
                                  test_dataloader=test_dataloader,
                                  optimizer=optimizer,
                                  loss_fn=loss_fn,
                                  epochs=10,
                                  device=device)
# End the timer and print out how long it took
end_time = timer()
print(f"[INFO] Total training time: {end_time-start_time:.3f} seconds")
```

### Make predictions on the entire test dataset with the model

```python
from tqdm.auto import tqdm
test_preds = []
base_model.eval()
with torch.inference_mode():
  for X, y in tqdm(test_dataloader):
    X, y = X.to(device), y.to(device)
    test_logits = base_model(X)
    pred_probs = torch.softmax(test_logits, dim=1)
    pred_label = torch.argmax(pred_probs, dim=1)
    test_preds.append(pred_label)

test_preds = torch.cat(test_preds).cpu()
test_preds
```

### Make a confusion matrix with the test preds and the truth labels

```python
test_truth = torch.cat([y for X,y in(test_dataloader)])
test_truth
```

## 2. Get the "most wrong" of the predictions on the test dataset and plot the 5 "most wrong" images. You can do this by:



* Predicting across all of the test dataset, storing the labels and predicted probabilities.
* Sort the predictions by *wrong prediction* and then *descending predicted probabilities*, this will give you the wrong predictions with the *highest* prediction probabilities, in other words, the "most wrong".
* Plot the top 5 "most wrong" images, why do you think the model got these wrong?

You'll want to:
* Create a DataFrame with sample, label, prediction, pred prob
* Sort DataFrame by correct (does label == prediction)
* Sort DataFrame by pred prob (descending)
* Plot the top 5 "most wrong" image predictions

```python
from pathlib import Path
test_data_path = list(Path(test_dir).glob("*/*.jpg"))
test_labels = [path.parent.stem for path in test_data_path]
test_labels
```

```python
def pred_and_keep(test_paths, model, transform, class_names,device):
  test_pred_list = []
  for path in tqdm(test_paths):
    pred_dict={}
    pred_dict["image_path"] = path
    class_name = path.parent.stem
    pred_dict["class_names"] = class_name

    from PIL import Image
    # Open image
    img = Image.open(path)
    transformed_image = transform(img).unsqueeze(0)
    model.eval()
    with torch.inference_mode():

      pred_logit = model(transformed_image.to(device))
      pred_prob = torch.softmax(pred_logit, dim=1)
      pred_label = torch.argmax(pred_prob, dim=1)
      pred_class = class_names[pred_label.cpu()]

      pred_dict["pred_prob"] = pred_prob.unsqueeze(0).max().cpu()
      pred_dict["pred_class"] = pred_class

      pred_dict["correct_label?"] = class_names == pred_class
      test_pred_list.append(pred_dict)

  return test_pred_list
```

```python
test_pred_example = pred_and_keep(test_paths=test_data_path,
                                  model=base_model,
                                  transform=transform_manually,
                                  class_names=class_names,
                                  device=device)
```

```python
test_pred_example[:5]
```

```python
# Convert this predictions into DataFrames for matplotlib
import pandas as pd
test_pred_df = pd.DataFrame(test_pred_example)
test_pred_df["pred_prob"]
```

```python
# Turn the test_pred_dicts into a DataFrame
import pandas as pd
test_pred_df = pd.DataFrame(test_pred_example)
# Sort DataFrame by correct then by pred_prob
top_5_most_wrong = test_pred_df.sort_values(by=["correct_label?", "pred_prob"], ascending=[True, False]).head()
top_5_most_wrong
```

```python
import matplotlib.pyplot as plt
import torchvision
for row in top_5_most_wrong.iterrows():
  row = row[1]
  image_path = row[0]
  true_label = row[1]
  pred_prob = row[2]
  pred_class = row[3]

  img = torchvision.io.read_image(str(image_path))
  plt.figure()
  plt.imshow(img.permute(1, 2, 0))
  plt.title(f"True: {true_label} | Pred: {pred_class} | Prob: {pred_prob:.3f}")
  plt.axis(False);
```
